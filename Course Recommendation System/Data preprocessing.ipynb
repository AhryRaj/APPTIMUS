```python
import pandas as pd
```

```python
# Load datasets
courses = pd.read_csv('courses.csv')
enrollments = pd.read_csv('enrollments.csv')
purchase_count = pd.read_csv('purchase_count.csv')
```

```python
# selecting some columns to form new dataframe
courses_cleaned = courses[['id','title', 'short_description']]
enrollments_cleaned = enrollments[['student_id','course_id','paid_amount']]

#renaming the column
courses_cleaned.rename(columns={'id': 'course_id'}, inplace=True)
```

```python
courses_cleaned.head(5)
```

```python
enrollments_cleaned.head(5)
```

```python
#creating a new column
enrollments_cleaned['is_purchased'] = enrollments_cleaned['paid_amount'].apply(lambda x: 0 if x == 0 else 1)
```

```python
#check whether there are any null values
enrollments_cleaned.isna().sum()
```

```python
courses_cleaned.isna().sum()
```

```python
# Handling missing values
enrollments_cleaned = enrollments_cleaned.dropna(subset=['student_id'])
```

```python
enrollments_cleaned.isna().sum()
```

```python
#changing data type
enrollments_cleaned['student_id'] = enrollments_cleaned['student_id'].astype(int)
```

```python
enrollments_cleaned.info()
```

```python
# Save the cleaned files to new CSV files
courses_cleaned.to_csv('courses_cleaned.csv', index=False)
enrollments_cleaned.to_csv('enrollments_cleaned.csv', index=False)
```
